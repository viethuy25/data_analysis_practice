```python
total_counts = bikes_2022.start_station_name.value_counts() # Value_counts ALREADY DONE SORTING FOR US

# Calculate cumulative percentages
cumulative_percentages = total_counts.cumsum() / total_counts.sum()
print ("Cumm per: ", cumulative_percentages)
print()

# Find the value at the 80th percentile of cumulative percentages
cutoff_percentile = np.percentile(cumulative_percentages, 80)

# Find the corresponding count cutoff value
cutoff_station = total_counts[cumulative_percentages <= cutoff_percentile].index[-1]
cutoff_count = total_counts[cumulative_percentages <= cutoff_percentile]

print(f"Cutoff station: {cutoff_station}")
print() 
print(f"Cutoff count: {cutoff_count}")
```

```python
plt.plot(cutoff_count)
plt.show()
```

The graph does not meant to show all station name, it is used for seeing the general trend. Here, it seems to be an exponential function

```python
from dateutil.parser import parse
```

```python
bikes_2022['started_at'] = bikes_2022['started_at'].apply(lambda x: parse(x))
bikes_2022['ended_at'] = bikes_2022['ended_at'].apply(lambda x: parse(x))

bikes_2022
```

```python
# calculate duration of leasing, denoted by minutes
# we ROUND UP because even crossing 1 SECOND into the next minute SHOULD BE CALCULATED 
import math

bikes_2022['duration'] = (bikes_2022['ended_at'] - bikes_2022['started_at']).apply(lambda x: math.ceil(float(x.total_seconds() / 60)))

bikes_2022
```

```python
bikes_2022.duration.value_counts()
```

```python
total_minute_rent = bikes_2022.duration.value_counts() # Value_counts ALREADY DONE SORTING FOR US

# Calculate cumulative percentages minutes
cumulative_percentages_minute = total_minute_rent.cumsum() / total_minute_rent.sum()
print ("Cumm percen min: ", cumulative_percentages_minute)
print()

# Find the value at the 80th percentile of cumulative percentages
cutoff_percentile_minute = np.percentile(cumulative_percentages_minute, 80)

# Find the corresponding count cutoff value
cutoff_minute = total_minute_rent[cumulative_percentages_minute <= cutoff_percentile_minute].index[-1]
cutoff_count_minute = total_minute_rent[cumulative_percentages_minute <= cutoff_percentile_minute]

print(f"Cutoff minute: {cutoff_minute}")
print() 
print(f"Cutoff count minute: {cutoff_count_minute}")
```

```python
plt.plot(cutoff_count_minute)
plt.show()
```

As can be seen, the majority of leasing time are under 100 minutes, with many outliers. Let's cut down further

```python
# Find the value at the top 20th percentile of cumulative percentages
twentyth_percentile_minute = np.percentile(cumulative_percentages_minute, 20)

# Find the corresponding count cutoff value
cutoff_twentyth_minute = total_minute_rent[cumulative_percentages_minute <= twentyth_percentile_minute].index[-1]
cutoff_twentyth_count_minute = total_minute_rent[cumulative_percentages_minute <= twentyth_percentile_minute]

print(f"Cutoff minute: {cutoff_twentyth_minute}")
print() 
print(f"Cutoff count minute: {cutoff_twentyth_count_minute}")
```

```python
plt.plot(cutoff_twentyth_count_minute)
plt.show()
```

```python
# Find the value at the top 5th percentile of cumulative percentages
fifth_percentile_minute = np.percentile(cumulative_percentages_minute, 5)

# Find the corresponding count cutoff value
cutoff_fifth_minute = total_minute_rent[cumulative_percentages_minute <= fifth_percentile_minute].index[-1]
cutoff_fifth_count_minute = total_minute_rent[cumulative_percentages_minute <= fifth_percentile_minute]

print(f"Cutoff minute: {cutoff_fifth_minute}")
print() 
print(f"Cutoff count minute: {cutoff_fifth_count_minute}")

plt.plot(cutoff_fifth_count_minute)
plt.show()
```

Even at the 20th and 5th top frequent leasing durations, there is still a significant drop off near the 100 minute. 

<b> Thus, the company should focus on pricing strategies below 100 minutes to maximize revenue and profit </b>

```python
# Convert 'started_at' column to datetime format
df = bikes_2022
df['started_at'] = pd.to_datetime(df['started_at'])

# Extract hour and day of the week from the 'started_at' column
df['hour'] = df['started_at'].dt.hour
df['day_of_week'] = df['started_at'].dt.day_name()

# Group by hour and day of the week and count the number of rides
grouped_data = df.groupby(['hour', 'day_of_week']).size().reset_index(name='ride_count')

#---------------------------------------------------------------------------------------------------------
# Define the order of the days PIVOT DATA SORT DAY BY DESIRED ORDER
desired_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


# Pivot the table for better visualization
pivot_data = grouped_data.pivot(index='hour', columns='day_of_week', values='ride_count')

# Reorder the columns based on the desired order
pivot_data = pivot_data[desired_order]
#-----------------------------------------------------------------------------------------------------

# Plotting the heatmap
plt.figure(figsize=(12, 8))
plt.imshow(pivot_data, cmap='viridis', aspect='auto')
plt.colorbar(label='Number of Rides')
plt.title('Number of Rides by Hour and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Hour of the Day')
plt.xticks(range(len(pivot_data.columns)), pivot_data.columns, rotation=45)
plt.yticks(range(len(pivot_data.index)), pivot_data.index)
plt.show()
```

```python
pivot_data
```

As can be seen from the graph above, people tends to rent more from 16:00 to 18:59 from Monday to Friday. During the weekend, it is spreed out more from 11:00 to 17:00.
